```python
import pandas as pd
```

```python
df_list_model = pd.read_csv('dokumentasi/evaluation_result_mei.csv')
df_list_model
```

```python
layer = []
unit = []
dropout = []
epoch = []
batch_size = []
for value in range(len(df_list_model)):
    curr = df_list_model['model_name'].loc[value]
    split = curr.split('_')
    batch_size.append(split[-1].split('.')[0])
    layer.append(split[1])
    unit.append(split[2])
    dropout.append(split[3])
    epoch.append(split[4])
df_list_model['layer'] = layer
df_list_model['unit'] = unit
df_list_model['dropout'] = dropout
df_list_model['epoch'] = epoch
df_list_model['batch_size'] = batch_size
```

```python
df_list_model.head(104)
```

```python
df_list_model = df_list_model[["stock", "model_name", "layer", "unit", "dropout", "epoch", "batch_size", "timestep", "mse", "mae", "rmse", "r2" ]]
```

```python
df_list_model.to_csv("dokumentasi/evaluation_result_right_ordered.csv")
```
